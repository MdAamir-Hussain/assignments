```python
#6.

import pandas as pd
import numpy as np

wine_quality = pd.read_csv('winequality-red.csv')
```

```python
from sklearn.decomposition import PCA

pca = PCA()

pca.fit(wine_quality)
```

```python
pca.explained_variance_ratio_
```

```python
#As you can see, the first two principal components explain over 43% of the variance in the data. The first three principal components explain over 62% of the variance, and the first four principal components explain over 76% of the variance.
 
```

```python
#To find the minimum number of principal components required to explain 90% of the variance in the data, we can use the following code
```

```python
np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.9)
```
