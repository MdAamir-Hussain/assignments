```python
#9..

import pandas as pd

# Load the Google Playstore dataset
df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Get the number of rows and columns
rows, columns = df.shape
print("Number of rows:", rows)
print("Number of columns:", columns)
```

```python
#10..

import pandas as pd
import matplotlib.pyplot as plt
```

```python
category_ratings = df.groupby('Category')['Rating'].mean()
```

```python
sorted_categories = category_ratings.sort_values(ascending=False).index
```

```python
sorted_df = pd.DataFrame({'Category': sorted_categories, 'Rating': category_ratings[sorted_categories]})
```

```python
plt.figure(figsize=(12, 6))
plt.boxplot(sorted_df['Rating'], vert=False)
plt.xlabel('Rating')
plt.ylabel('App Category')
plt.title('Rating Distribution by App Category')
plt.show()
```

```python
filtered_df = df.dropna(subset=['Size', 'Rating'])
```

```python
filtered_df['Size'] = filtered_df['Size'].str.replace('M', '').str.replace('k', '').astype(float)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Size'], filtered_df['Rating'])
plt.xlabel('App Size')
plt.ylabel('Rating')
plt.title('Relationship between App Size and Rating')
plt.show()
```

```python
#13.
import pandas as pd
import matplotlib.pyplot as plt

# Filter out non-numeric prices and convert to numeric format
filtered_df = df[df['Price'].apply(lambda x: str(x).replace('$', '').replace('Everyone', '').isnumeric())]
filtered_df['Price'] = pd.to_numeric(filtered_df['Price'].str.replace('$', ''), errors='coerce')

# Calculate average prices by app type
type_prices = filtered_df.groupby('Type')['Price'].mean()

# Create bar chart
plt.figure(figsize=(10, 6))
type_prices.plot(kind='bar', color='blue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Price by App Type')
plt.xticks(rotation=0)
plt.show()
```
